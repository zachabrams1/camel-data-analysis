```python
import pandas as pd
import os

dfs = {}
for file in os.listdir("./Raw"):
    if file.endswith(".csv"):
        dfs[file.split(".csv")[0]] = pd.read_csv(f"./Raw/{file}")
```

```python
dfs_sheets = {}
for key in dfs:
    if "Sheets" in key:
        dfs_sheets[key] = dfs[key]

dfs_partiful =  {}
for key in dfs:
    if "Partiful" in key:
        dfs_partiful[key] = dfs[key]

dfs_posh = {}
for key in dfs:
    if "Posh" in key:
        dfs_posh[key] = dfs[key]

dfs_others = {}
for key in dfs:
    if key not in dfs_sheets and key not in dfs_partiful and key not in dfs_posh:
        dfs_others[key] = dfs[key]

# print lengths of each dictionary
print("Length of dfs: ", len(dfs))
print("Length of dfs_sheets: ", len(dfs_sheets))
print("Length of dfs_partiful: ", len(dfs_partiful))
print("Length of dfs_posh: ", len(dfs_posh))
print("Length of dfs_others: ", len(dfs_others))

# track totals observations
total_len = 0
for df in dfs.values():
    total_len += len(df)
```

```python
from fuzzywuzzy import fuzz 
def compute_match_score(rowA, rowB):
    score = 0.0
    
    # auto merge if emails are the same
    if pd.notna(rowA["Email"]) and pd.notna(rowB["Email"]):
        if rowA["Email"] == rowB["Email"]:
            score += 50.0

    # 1. Fuzzy match first names
    if pd.notna(rowA["First Name"]) and pd.notna(rowB["First Name"]):
        ratio = fuzz.ratio(str(rowA["First Name"]).lower(), str(rowB["First Name"]).lower())
        # If ratio is above 90 => 1 point, above 80 => 0.5 point, else 0
        if ratio > 90:
            score += 1.0
    
    # 2. Exact match last names
    if pd.notna(rowA["Last Name"]) and pd.notna(rowB["Last Name"]):
        if rowA["Last Name"].strip().lower() == rowB["Last Name"].strip().lower():
            score += 1.0
        else:
            score -= 2.0
    
    # 3. Same school
    if pd.notna(rowA["School"]) and pd.notna(rowB["School"]):
        if rowA["School"].lower() == rowB["School"].lower():
            score += 1.0
        else:
            score -= 2.0

    # 4. Same grade
    if pd.notna(rowA["Grade"]) and pd.notna(rowB["Grade"]):
        if rowA["Grade"] == rowB["Grade"]:
            score += 1.0
        else:
            score -= 1.0
    
    # 5. Same gender
    if pd.notna(rowA["Gender"]) and pd.notna(rowB["Gender"]):
        if rowA["Gender"] == rowB["Gender"]:
            score += 0.5
        else:
            score -= 2.0

    # 6. Same is_jewish
    if pd.notna(rowA["Is_Jewish"]) and pd.notna(rowB["Is_Jewish"]):
        if rowA["Is_Jewish"] == rowB["Is_Jewish"]:
            score += 1.0
        else:
            score -= 0.5
    
    return score

def find_duplicates(df, score_threshold=1.0):
    """
    Returns a list of tuples (i, j, score) for likely duplicates.
    We'll do an O(n^2) approach here for simplicity.
    """
    potential_dupes = []
    for i in range(len(df)):
        for j in range(i+1, len(df)):
            rowA = df.iloc[i]
            rowB = df.iloc[j]
            s = compute_match_score(rowA, rowB)
            if s >= score_threshold:  # Only return if the score is at least 1
                potential_dupes.append((i, j, s))
    return potential_dupes


def merge_records(rowA, rowB):
    """
    Return a new 'merged' Series where we keep the best info from rowA and rowB.
    For emails, we create a list of unique emails.
    """
    merged = {}
    
    # Merge first name - if rowA has data, keep it; otherwise use rowB
    # Or you can pick the "longer" or "more complete" name, etc.
    merged["First Name"] = rowA["First Name"] if pd.notna(rowA["First Name"]) else rowB["First Name"]
    
    # Similarly for last name
    merged["Last Name"] = rowA["Last Name"] if pd.notna(rowA["Last Name"]) else rowB["Last Name"]
    
    # School, Grade, Gender, etc. - same pattern
    merged["School"] = rowA["School"] if pd.notna(rowA["School"]) else rowB["School"]
    merged["Grade"]  = rowA["Grade"]  if pd.notna(rowA["Grade"])  else rowB["Grade"]
    merged["Gender"] = rowA["Gender"] if pd.notna(rowA["Gender"]) else rowB["Gender"]
    
    # For is_jewish, pick one if the other is None, or if they match. 
    if pd.isna(rowA["Is_Jewish"]):
        merged["Is_Jewish"] = rowB["Is_Jewish"]
    else:
        merged["Is_Jewish"] = rowA["Is_Jewish"]
    
    # Combine emails into a list
    emails = set()
    if pd.notna(rowA["Email"]):
        emails.add(rowA["Email"])
    if pd.notna(rowB["Email"]):
        emails.add(rowB["Email"])
        
    merged["Email"] = list(emails)  # store as a list
    
    return pd.Series(merged)
```

```python
def process_duplicates(df, auto_merge_threshold=2.5, manual_review_threshold=2.0):
    """
    - If score >= auto_merge_threshold => auto-merge
    - Else if score >= manual_review_threshold => ask user to confirm
    - Below manual_review_threshold => ignore
    """
    dupes = find_duplicates(df, score_threshold=manual_review_threshold)
    
    # We'll store merges in a list of (i, j, merged_row)
    merges_to_perform = []
    rows_to_drop = set()
    
    for (i, j, score) in dupes:
        if i in rows_to_drop or j in rows_to_drop:
            # Already merged or dropped, skip
            continue
        
        if score >= auto_merge_threshold:
            # Auto-merge
            merged_row = merge_records(df.loc[i], df.loc[j])
            merges_to_perform.append((i, j, merged_row))
        else:
            # Prompt user
            print(f"\nPossible duplicate (score={score:.2f}):")
            print("Row i:\n", df.loc[i])
            print("Row j:\n", df.loc[j])
            choice = input("Merge these? [y/n] ")
            if choice.lower().startswith("y"):
                merged_row = merge_records(df.loc[i], df.loc[j])
                merges_to_perform.append((i, j, merged_row))

    # Now actually perform merges
    # For each pair, we create a new row in the DF (or update one of them),
    # then mark the other as dropped
    for (i, j, merged_row) in merges_to_perform:
        # Put the merged row in the place of i, and drop j
        df.loc[i] = merged_row
        rows_to_drop.add(j)
    
    # Drop the merged rows
    df.drop(rows_to_drop, inplace=True)
    
    # Re-index if you like
    df.reset_index(drop=True, inplace=True)
    
    return df

# Example usage
df_merged = process_duplicates(df_all_people)
```
