```python
import pandas as pd

people, contacts, attendance = pd.read_csv("final/people.csv"), pd.read_csv("final/contacts.csv"), pd.read_csv("final/attendance.csv")
```

```python
people = people.merge(attendance.groupby("person_id").size().reset_index().rename(columns={0: "attendance_count"}), left_on="id", right_on="person_id").drop(columns=["person_id"])
```

```python
contacts_piv = contacts.pivot_table(
    index="person_id",
    columns="contact_type",
    values="contact_value",
    aggfunc="first"  # in case there are duplicates
).reset_index()

# Optional: rename columns if needed
contacts_piv = contacts_piv.rename(columns={
    "school email": "school_email",
    "personal email": "personal_email"
})
people = people.merge(contacts_piv, left_on="id", right_on="person_id").drop(columns=["person_id"])
```

```python
people.sort_values("attendance_count", ascending=False).to_csv("robbie_export.csv")
```
